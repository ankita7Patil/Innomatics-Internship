# Problem Statement 1: Employee Performance Bonus Eligibility


```python
# Employee Bonus Eligibility Program

staff_scores = {
    "Arjun": 92,
    "Neha": 88,
    "Rahul": 92,
    "Meera": 85
}

# determine highest score
top_score = max(staff_scores.values())

# collect employees with highest score
bonus_candidates = []

for employee, marks in staff_scores.items():
    if marks == top_score:
        bonus_candidates.append(employee)

# display result
print("Top Performers Eligible for Bonus:",
      ", ".join(bonus_candidates),
      f"(Score: {top_score})")
```

# Problem Statement 2: Search Query Keyword Analysis

```python
def keyword_analysis(sentence):
    text = sentence.lower()

    for ch in [".", ",", "!", "?", ";", ":"]:
        text = text.replace(ch, "")

    words = text.split()
    freq = {}

    for word in words:
        freq[word] = freq.get(word, 0) + 1

    result = {}
    for k, v in freq.items():
        if v > 1:
            result[k] = v

    return result


query = "Buy mobile phone buy phone online"
print(keyword_analysis(query))
```

# Problem Statement 3: Sensor Data Validation

```python
def validate_sensor_data(readings):
    valid_pairs = []

    for i in range(len(readings)):
        value = readings[i]
        if value % 2 == 0:
            valid_pairs.append((i, value))

    return valid_pairs


sensor_readings = [3, 4, 7, 8, 10, 12, 5]

print("Valid Sensor Readings (Hour, Value):")
print(validate_sensor_data(sensor_readings))
```

# Problem Statement 4: Email Domain Usage Analysis

```python
def domain_usage(email_list):
    usage = {}

    for mail in email_list:
        dom = mail.split("@")[1]
        usage[dom] = usage.get(dom, 0) + 1

    total = len(email_list)

    for dom, cnt in usage.items():
        pct = (cnt / total) * 100
        print(f"{dom}: {pct:.0f}%")


emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_usage(emails)
```

# Problem Statement 5: Sales Spike Detection

```python


daily_revenue = [1200, 1500, 900, 2200, 1400, 3000]

total_revenue = 0
for value in daily_revenue:
    total_revenue += value

avg_revenue = total_revenue / len(daily_revenue)


spike_limit = avg_revenue * 1.30


for day_index in range(len(daily_revenue)):
    sale_value = daily_revenue[day_index]

    if sale_value > spike_limit:
        print(f"Day {day_index + 1}: {sale_value}")
```

# Problem Statement 6: Duplicate User ID Detection

```python
def find_duplicate_ids(id_list):
    freq_map = {}

    for item in id_list:
        freq_map[item] = freq_map.get(item, 0) + 1

    for key, val in freq_map.items():
        if val > 1:
            print(f"{key} → {val} times")


registered_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

find_duplicate_ids(registered_ids)
```

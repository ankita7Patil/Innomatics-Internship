# Problem 1: Mobile Recharge Validation System

```python
def recharge_check():
    allowed_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Amount must be at least ₹50")
        elif amount not in allowed_plans:
            print("Invalid plan. Choose from 199, 299, 399, 599")
        else:
            print("Recharge Successful ")
            break

recharge_check()
```

# Problem 2: Inventory Reorder Alert System

```python
def stock_monitor(product_data):
    for item, qty in product_data.items():
        if qty < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")


warehouse_stock = {
    "Pens": 12,
    "Notebooks": 25,
    "Markers": 8,
    "Staplers": 18
}

stock_monitor(warehouse_stock)
```

# Problem 3: Student Result Processing System

```python
def result_status(score_list):
    total_marks = 0

    for marks in score_list:
        total_marks += marks

    avg_marks = total_marks / len(score_list)

    if avg_marks >= 50:
        return "Pass"
    else:
        return "Fail"


student_scores = [45, 60, 55, 70]

print("Result:", result_status(student_scores))
```

# Problem 4: Cab Fare Estimator with Retry

```python
def cab_fare():
    base = 50
    rate = 12

    while True:
        distance = float(input("Enter distance in km: "))
        peak = input("Peak hour? (yes/no): ").lower()

        fare = base + (distance * rate)

        if peak == "yes":
            fare = fare * 1.25

        print("Estimated Fare: ₹", round(fare, 2))

        again = input("Calculate again? (yes/no): ").lower()
        if again != "yes":
            break

cab_fare()
```

# Problem 5: Employee Attendance Eligibility Checker

```python
def attendance_check(record):
    present_days = 0

    for status in record:
        if status == "P":
            present_days += 1

    percent = (present_days / len(record)) * 100

    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


emp_attendance = ["P", "P", "A", "P", "P", "A", "P"]

print("Attendance Status:", attendance_check(emp_attendance))
```

# Problem 6: Password Strength Checker

```python
def password_strength(pwd):
    has_digit = False
    has_special = False

    if len(pwd) < 8:
        return "Weak Password"

    for ch in pwd:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


user_pass = input("Enter password: ")
print(password_strength(user_pass))
```

# 1.Logic Building Task




```python
def check_login():
    print("\n========== LOGIN VERIFICATION ==========")

    store_username = "admin"
    store_password = "1234"

    user_input_name = input("Enter Username: ")
    user_input_pass = input("Enter Password: ")

    if user_input_name == store_username:
        if user_input_pass == store_password:
            print("Login Successful")
        else:
            print("Password Incorrect")
    else:
        print("Username Not Found")
```

# 2. Pass / Fail Analyzer


```python

def evaluate_results():
    print("\n========== STUDENT RESULT ANALYSIS ==========")

    student_scores = [45, 78, 90, 33, 60]

    passed = 0
    failed = 0

    for score in student_scores:
        if score >= 50:
            passed += 1
        else:
            failed += 1

    print("Total Students:", len(student_scores))
    print("Number of Students Passed:", passed)
    print("Number of Students Failed:", failed)
```

# 3. Simple Data Cleaner

```python
def process_names():
    print("\n========== DATA CLEANING PROCESS ==========")

    name_list = [" Alice ", "bob", " CHARLIE "]
    updated_list = []

    for item in name_list:
        cleaned_item = item.strip()      # remove extra spaces
        cleaned_item = cleaned_item.lower()  # convert to lowercase
        updated_list.append(cleaned_item)

    print("Original List:", name_list)
    print("Cleaned List:", updated_list)
```

# 4. Message Length Analyzer

```python
def analyze_messages():
    print("\n========== MESSAGE LENGTH CHECK ==========")

    message_data = ["Hi", "Welcome to the platform", "OK"]

    for text in message_data:
        message_size = len(text)
        print(f"Message: '{text}' → Length: {message_size}")

        if message_size > 10:
            print("This message exceeds the recommended length")
```

# 5. Error Message Detector


```python
def analyze_logs():
    print("\n========== SYSTEM LOG ANALYSIS ==========")

    log_entries = ["INFO", "ERROR", "WARNING", "ERROR"]

    error_counter = 0

    for log in log_entries:
        if log == "ERROR":
            error_counter += 1

    print("Total Log Records:", len(log_entries))
    print("Total ERROR Records:", error_counter)
```

```python
check_login()
evaluate_results()
process_names()
analyze_messages()
analyze_logs()
```

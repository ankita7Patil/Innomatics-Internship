 # Problem Statement 1:Smart Parking Lot Management System


```python
def track_parking(limit, activity_log):
    current_count = 0
    highest_load = 0

    for move in activity_log:
        if move == "IN":
            current_count += 1
        elif move == "OUT" and current_count > 0:
            current_count -= 1

        if current_count > highest_load:
            highest_load = current_count

    state = "Capacity Reached" if current_count >= limit else "Slots Available"

    print("Currently Parked Vehicles:", current_count)
    print("Parking Status:", state)
    print("Peak Parking Load:", highest_load)


track_parking(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

#  Problem Statement 2:Online Food Delivery Time Estimator


```python
def delivery_eta(distance, traffic, weather):
    time = distance * 5   # base 5 min per km

    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8

    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


delivery_eta(8, "High", "Rainy")
```

# Problem 3: Theatre Seat Occupancy

```python
def theatre_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    percent = (booked / total_seats) * 100

    if percent == 100:
        state = "Housefull"
    elif percent >= 70:
        state = "Almost Full"
    else:
        state = "Seats Available"

    print("Occupancy:", round(percent), "%")
    print("Show Status:", state)


theatre_occupancy(200, [1]*150)
```

# Problem 4: Cloud Server Load Classification

```python
def server_load(cpu_values):
    total = 0
    for v in cpu_values:
        total += v

    avg = total / len(cpu_values)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", round(avg), "%")
    print("Server Status:", status)


server_load([45,60,70,85,90])
```

# Problem 5: Classroom Resource Monitor

```python
def resource_usage(resource_hours):
    overused = []

    for name, hrs in resource_hours.items():
        if hrs > 8:
            overused.append(name)

    alert = "Yes" if overused else "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", alert)


usage_data = {"Projector":6, "AC":9, "Lights":4}
resource_usage(usage_data)
```

# Problem 6: Event Registration Controller

```python
def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


event_registration(100, 105)
```

# Problem 1: Smart Light Controller

```python
class LightUnit:
    def __init__(self, label):
        self.label = label
        self.state = "OFF"

    def switch(self, action):
        self.state = action.upper()

    def show_status(self):
        print(f"{self.label} is {self.state}")


lamp = LightUnit("Bedroom Light")
lamp.switch("ON")
lamp.show_status()
```

# Problem 2: Employee ID Card System

```python
class StaffCard:
    def __init__(self, name, emp_code, dept):
        self.name = name
        self.emp_code = emp_code
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_code)
        print("Department:", self.dept)


emp = StaffCard("Rahul", "EMP102", "IT")
emp.display()
```

# Problem 3: Mobile Contact Record

```python
class ContactEntry:
    def __init__(self, person, phone):
        self.person = person
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.person)
        print("Phone:", self.phone)


c1 = ContactEntry("Anita", "9876543210")
c1.show_contact()
```

# Problem 4: Product Price Tag Generator

```python
class ItemTag:
    def __init__(self, item_name, cost):
        self.item_name = item_name
        self.cost = cost

    def print_tag(self):
        print("Product:", self.item_name)
        print("Price: ₹" + str(self.cost))


p1 = ItemTag("Headphones", 2499)
p1.print_tag()
```

# Problem 5: Movie Rating Display System

```python
class FilmInfo:
    def __init__(self, title, score):
        self.title = title
        self.score = score

    def display_rating(self):
        print("Movie:", self.title)
        print("Rating:", self.score, "/ 5")


m1 = FilmInfo("Inception", 4.8)
m1.display_rating()
```

# Problem 6: Delivery Address Manager

```python
class DeliveryInfo:
    def __init__(self, customer, location):
        self.customer = customer
        self.location = location

    def show_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.location)


d1 = DeliveryInfo("Suman", "Hyderabad")
d1.show_delivery()
```

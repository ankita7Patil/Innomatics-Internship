# Problem 1: Post Engagement Analyzer

```python
def engagement_check(like_list):
    total_likes = 0
    for l in like_list:
        total_likes += l

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


post_likes = [120, 300, 250, 380]
engagement_check(post_likes)
```

# Problem 2: Medicine Stock Alert

```python
def medicine_stock(level):
    if level < 10:
        state = "Low Stock Alert"
    else:
        state = "Stock Sufficient"

    print("Medicine Stock:", level)
    print("Status:", state)


medicine_stock(6)
```

# Problem 3: Rainfall Adequacy Checker

```python
def rainfall_check(rain_data, required):
    total = 0
    for r in rain_data:
        total += r

    avg = total / len(rain_data)

    result = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"

    print("Average Rainfall:", avg)
    print("Rainfall Status:", result)


season_rain = [70, 80, 65, 73]
rainfall_check(season_rain, 70)
```

# Problem 4: Duplicate Account Detection

```python
def duplicate_users(usernames):
    if len(usernames) != len(set(usernames)):
        flag = "Yes"
    else:
        flag = "No"

    print("Duplicate Accounts Found:", flag)


accounts = ["userA", "userB", "userA", "userC"]
duplicate_users(accounts)
```

# Problem 5: Appointment Eligibility

```python
def appointment_age(age):
    status = "Eligible" if age >= 18 else "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)


appointment_age(21)
```

# Problem 6: Premium Crop Filter

```python
def premium_crops(price_list):
    premium = []

    for p in price_list:
        if p > 2000:
            premium.append(p)

    print("Premium Crops:", premium)


crop_prices = [1500, 2500, 1800, 3200]
premium_crops(crop_prices)
```

# Problem 7: Application Health Checker

```python
def system_health(errors):
    if errors == 0:
        state = "Healthy"
    elif errors <= 5:
        state = "Minor Issues"
    else:
        state = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", state)


system_health(7)
```

# Problem 8: Transaction Limit Checker

```python
def transaction_check(amount):
    limit = 50000

    status = "Approved" if amount <= limit else "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


transaction_check(60000)
```

# Problem 9: Attendance Eligibility

```python
def attendance_exam(record):
    present = 0

    for r in record:
        if r == "P":
            present += 1

    percent = (present / len(record)) * 100
    eligibility = "Eligible" if percent >= 75 else "Not Eligible"

    print("Attendance Percentage:", percent)
    print("Exam Eligibility:", eligibility)


attendance_data = ["P", "P", "A", "P", "P"]
attendance_exam(attendance_data)
```

# Problem 10: Smart Electricity Bill

```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = 100*3 + (units-100)*5
    else:
        bill = 100*3 + 100*5 + (units-200)*7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Level:", usage)


electricity_bill(280)
```
